```python
from IPython.display import clear_output

class Cart:
    items = []

    def add_item(self, item):
        return self.items.append(item)
    
    def remove_item(self, rem_item):
        for a in self.items:
            if rem_item in [i.name for i in self.items]:
                self.items.remove(a)
       
    def clear_items(self):
        self.items.clear()
     
    def show_items(self):
        total = 0
        if not self.items:
            print("You have no items in your cart. ")
        else:
            display_cart = []
            for p in self.items:
                if p.name not in [i['name'] for i in display_cart]:
                    display_cart.append({
                        'name': p.name,
                        'price': p.price
                    })
                else:
                    for i in display_cart:
                        if i['name'] == p.name:
                            i['price'] += p.price

            for i in range(len(display_cart)):
                print(f"{i+1}: {display_cart[i]['name']} [{[i.name for i in self.items].count(display_cart[i]['name'])}] ${display_cart[i]['price']}")

        total += float(sum([p.price for p in self.items]))

        print(f"Total: ${total:,.2f}")
    
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class ShoppingCart:
    def instructions():
        print("Type 'add' to add an item, 'remove' to remove an item, 'show' to view your cart, 'clear' to remove all items or 'quit' to exit the program.")
        print('-' * 50)
            
    @classmethod
    def run(self):
        cart = Cart()
        
        flag = False
        while flag is not True:
            clear_output()
            self.instructions()
            cart.show_items()
            
            ask = input("What would you like to do? ").lower()
            if ask == 'add':
                item_name = input("What item would you like to add? ")
                price = float(input("How much does the item cost? ")) 
                cart_item = Item(item_name, price)
                cart.add_item(cart_item)
            elif ask == 'clear':
                cart.clear_items()         
            elif ask == 'remove':
                cart_item = input("What would you like to remove? ").lower()
                cart.remove_item(cart_item)
            elif ask == 'quit':
                question = input("Are you sure you want to quit? Y/N: ")
                if question == 'y':
                    print("Thank you for shopping, have a great day!")
                    flag = True
                if question == 'n':
                    break
            elif ask == 'show':
                cart.show_items() 
            else:
                input("Invalid input, press 'ENTER' to continue. ")
            
ShoppingCart.run()
```
